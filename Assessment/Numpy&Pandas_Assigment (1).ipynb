```python
#13. What are the features of TensorFlow?
TensorFlow is a powerful open-source machine learning framework developed by Google. It provides a comprehensive ecosystem for building and deploying machine learning and deep learning models. Here are the key features of TensorFlow:
1. Open-Source and Cross-Platform:
    Free to use and actively maintained by Google.
    Works on Windows, macOS, Linux, and mobile platforms (Android, iOS).
    Supports CPU, GPU, and TPU acceleration.
2. Computation Graphs:
    Uses data flow graphs where nodes represent operations and edges represent tensors.
    Allows easy deployment to multiple devices.
3. Eager Execution:
    TensorFlow 2.x introduced eager execution by default.
    Enables intuitive and flexible coding (like NumPy), with immediate results.
4. Rich APIs:
    High-level API: Keras (for fast prototyping and model building).
    Low-level API: Direct control for custom model architecture and training loops.
5. TensorBoard
    Built-in tool for visualizing:
        Training progress
        Metrics (loss, accuracy)
        Graph structures
        Histograms, images, and more
6. Deployment Options
    TensorFlow Lite – for mobile and embedded devices.
    TensorFlow.js – for running models in the browser.
    TensorFlow Serving – for model deployment in production environments.
7. Wide Range of ML Support
    Supports both traditional ML and deep learning:
        Neural networks
        CNNs, RNNs, Transformers
        Reinforcement learning
        Time-series forecasting, etc.
8. Pre-trained Models & Model Hub
    Access to TensorFlow Hub for pre-trained models.
    Easy to reuse models for transfer learning.
9. AutoML & Estimators
    Tools for automatic model tuning and scalable training.
    tf.estimator API for distributed training.
10. Integration with Other Tools
    Easily integrates with NumPy, Pandas, scikit-learn, and more.
    Can be combined with TPUs and GPUs for massive acceleration
```

```python
#14. List a few limitations of TensorFlow.
Here are some limitations of TensorFlow, despite its powerful features:
1. Steep Learning Curve:
    TensorFlow (especially before v2) can be complex and less intuitive for beginners compared to frameworks like PyTorch.
    The use of computation graphs (in TensorFlow 1.x) was hard to debug.
2. Debugging Challenges:
    While TensorFlow 2.x has improved with eager execution, low-level operations and silent errors can still make debugging difficult.
3. Verbose Syntax:
    Writing custom layers or training loops can require more boilerplate code than PyTorch.
4. Limited Dynamic Computation (vs PyTorch):
    TensorFlow was initially designed for static computation graphs, which made dynamic models (e.g., RNNs with variable input sizes) harder to implement naturally. Although this is better in TF 2.x, PyTorch still has more flexibility.
5. Less Pythonic:
    TensorFlow’s style often deviates from standard Python practices, making it feel more like a separate domain-specific language.
6. Overhead and Complexity:
    The ecosystem is large and sometimes bloated, which can be overwhelming. Tools like TensorBoard, TFX, and Estimators are powerful but may be overkill for simple projects.
7. Compatibility Issues:
    Version mismatches (e.g., between TensorFlow and CUDA/cuDNN) can cause installation and runtime issues.
```

```python
#15. What do you know about supervised and unsupervised machine learning?
Supervised vs Unsupervised Machine Learning
Machine learning is typically categorized into supervised and unsupervised learning based on the type of data used during training.

Supervised Learning
    Definition:
        Supervised learning uses labeled data, meaning each input has a corresponding correct output.

    Key Characteristics:
        Learns a function that maps inputs to outputs.
        Requires a training dataset with input-output pairs.
        The model is "supervised" by the labels during training.

    Examples:
        Input (X)	Output (Y)
        Image of a cat	"Cat"
        House features	Price in dollars
        Email text	Spam or Not Spam

    Common Algorithms:
        Classification: Logistic Regression, Decision Trees, SVM, Neural Networks
        Regression: Linear Regression, Random Forest Regressor, Lasso, etc.

    Use Cases:
        Email spam detection
        Fraud detection
        House price prediction
        Medical diagnosis

Unsupervised Learning:=
    Definition:
        Unsupervised learning uses unlabeled data, so the algorithm tries to find hidden patterns or groupings without known outputs.

    Key Characteristics:
        No labels are provided—only input data.
        Goal is to discover structure, clusters, or associations in the data.

    Examples:
        Input (X)	Output (Y) (None)
        Customer data	(Find similar groups)
        Website activity logs	(Identify behavior patterns)

    Common Algorithms:
        Clustering: K-Means, DBSCAN, Hierarchical Clustering
        Dimensionality Reduction: PCA, t-SNE, Autoencoders
        Association Rules: Apriori, FP-Growth

    Use Cases:
        Customer segmentation
        Market basket analysis
        Anomaly detection
        Topic modeling (e.g., LDA)
```
